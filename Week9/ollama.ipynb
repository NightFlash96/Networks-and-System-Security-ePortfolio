```python
from ollama import chat, ChatResponse

response: ChatResponse = chat(model="tinyllama:1.1b", messages=[
    {
    "role": "user", 
    "content": "Hello, Ollama!"
    }
])
print(response["message"]["content"])
# or you can access fields directly from the reponse object
print(response.message.content)
```
